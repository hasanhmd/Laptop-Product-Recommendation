# B. Instantiate Data Context

```python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

# C. Connect to A `Datasource`

```python
# For load data in drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'milestone_3'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'laptop_clean'
path_to_data = '/content/drive/MyDrive/P2M3_hasan_abdul_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

# D. Create an Expectation Suite

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-laptop-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

## D.1 - Expectations

**We expect from the data.**

- laptop to be unique
- ram to be between 4 and 128
- status to be in New and Refurbished
- final_price to be in type float
- laptop value counts betweem 500 and 2500
- stdev of screen between 0 and 5
- final_price be average in 0 to 20000

*You can see the galery of Expectations [here](https://docs.greatexpectations.io/en/latest/reference/glossary_of_expectations.html)*

```python
# Expectation 1 : Column `laptop` must be unique

validator.expect_column_values_to_be_unique('laptop')
```

```python
# Expectation 2 : Column `ram` must be between 4 and 128

validator.expect_column_values_to_be_between(
    column='ram', min_value=4, max_value=128
)
```

```python
# Expectation 3 : Column `status` must contain one of the following 2 things :
'''
- New
- Refurbished
'''

validator.expect_column_values_to_be_in_set('status', ['New', 'Refurbished'])
```

```python
# Expectation 4 : Column `final_price` must in form of float

validator.expect_column_values_to_be_in_type_list('final_price', ['float'])
```

```python
# Expectation 5 : Column `laptop` count between 500 and 2500

validator.expect_table_column_count_to_be_between(column='laptop', min_value=500, max_value=2500)
```

```python
# Expectation 6 : Expect stdev of screen between 0 and 5

validator.expect_column_stdev_to_be_between(
    column = 'screen',
    min_value = 0,
    max_value = 5,
    strict_min = True,
    strict_max = True)
```

```python
# Expectation 7 : The average of `final_price` must in range 0 - 20000

validator.expect_column_mean_to_be_between('final_price', 0, 20000)
```

```python
# Save into Expectation Suite

validator.save_expectation_suite(discard_failed_expectations=False)
```

## D.2 - Checkpoint

```python
# Create a checkpoint

checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

```python
# Run a checkpoint

checkpoint_result = checkpoint_1.run()
```
